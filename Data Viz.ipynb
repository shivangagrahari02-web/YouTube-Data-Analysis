```python
import matplotlib.pyplot as plt
```

```python
channels = ["Adda247", "WiFiStudy", "Mahendra Guru", "T-Series Bhakti Sagar", 
            "Kori Pati Productions", "Study IQ education", "Vikram Aditya", 
            "Studio Saraswati Official", "zeetvtelugu", "T-Series"]
trending_videos = [51, 36, 16, 14, 12, 12, 10, 10, 7, 7]
```

```python
plt.figure(figsize=(12, 6))
plt.bar(channels, trending_videos, color='skyblue')
plt.xlabel('Channels')
plt.ylabel('Number of Videos Trending on the Same Day')
plt.title('Number of Videos Trending on the Same Day of Upload for Each Channel')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
months = ["March", "April", "January", "December", "February", "June", "November", "May"]
uploads = [44, 39, 31, 27, 27, 23, 16, 14]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(months, uploads, color='salmon')
plt.xlabel('Months')
plt.ylabel('Number of Videos Uploaded')
plt.title('Number of Videos Uploaded per Month by T-Series')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.pie(uploads, labels=months, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6','#c4e17f','#ff6666'])
plt.title('Number of Videos Uploaded per Month by T-Series')
plt.axis('equal')
plt.show()
```
